# Data Project 1: FIFA 20 Players 

```python
import pandas as pd 
import numpy as py 
import seaborn as sns
import matplotlib.pyplot as plt
```

1.) Load CSV with top 5 records

```python
fifa = pd.read_csv('players_20.csv')
fifa.head()
```

2. All Column Names

```python
for col in fifa.columns:
    print(col)
```

3.) Number of Rows and Columns 

```python
fifa.info()
```

4.) Number of Players from each Country 

```python
fifa['nationality'].value_counts()
```

5.) Top 10 Countries with highest number of Fifa players 

```python
fifa['nationality'].value_counts().head(10)
```

6.) Bar Chart that includes Top 5 Countries with highest number of players 

```python
player_count = fifa['nationality'].value_counts()[:5]
player_count
```

```python
plt.figure(figsize=(10,5))
sns.barplot(player_count.index, player_count.values, color = 'Green')
plt.title("Top 5 Countries with the Most Players", fontsize = 16)
plt.ylabel("Number of Players", fontsize = 12)
plt.xlabel("Countries", fontsize = 12)
plt.show()
```

7.) Top 5 players and Wages

```python
fifa[['short_name','wage_eur']].head() 
```

8.) Top 5 Players with Highest Wages

```python
player = fifa[['short_name','wage_eur']].sort_values(by = ['wage_eur'],ascending = False).head()
player
```

9.) Bar Chart Top 5 Players with Highest Wages

```python
plt.figure(figsize=(10,5))
sns.barplot(player['short_name'], player['wage_eur'])
plt.title("Top 5 players with Highest Wages", fontsize = 16)
plt.ylabel("Wage (Euros)", fontsize = 12)
plt.xlabel("Player", fontsize = 12)
plt.show()
```
