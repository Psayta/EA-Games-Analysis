```python
import pandas as pd 
import numpy as py 
import seaborn as sns
import matplotlib.pyplot as plt
import squarify as sq
```

```python
fifa = pd.read_csv('players_20.csv')
fifa
```

Right and Left foot comparison 

```python
foot = fifa['preferred_foot'].value_counts()
foot
```

```python
plt.figure(figsize=(10,5))
sns.color_palette("bright")
sns.barplot(foot.index, foot.values).axhline(4318, color = 'black')
plt.title("Right vs Left Foots", fontsize = 16)
plt.show()
```

Age Analysis 

```python
fifa['age'].describe()
```

```python
ages = fifa['age'].value_counts()
ages
```

```python
#Number of Players at each age 
plt.figure(figsize=(10,5))
sns.barplot(ages.index, ages.values)
```

```python
#Sprint speed regression overtime 
plt.figure(figsize=(10,5))
sns.regplot(x = 'age', y = 'movement_sprint_speed' , data = fifa, x_jitter=.4, scatter_kws={"color": "blue"}, line_kws={"color": "red"} )
```

```python
# Speed Averages for each age 
plt.figure(figsize=(10,5))
sns.barplot(fifa['age'], fifa['movement_sprint_speed'])
```

Overall Player Analysis

```python
plt.figure(figsize=(10,5))
sns.barplot(OVR.index, OVR.values)
```

Nationality Counts 

```python
a = fifa.groupby('nationality').count().sort_values('nationality').index.get_level_values(0).tolist()
d = fifa.groupby('nationality').count().sort_values('overall',ascending = False)
b = d['overall'].index
s = d['overall'].sort_values(ascending = False)
s
```

```python
fig, ax = plt.subplots(1, figsize = (15,15))
sq.plot(sizes=s, label=b, alpha=.8)
plt.axis('off')
plt.show()
```
